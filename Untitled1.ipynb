### HEART DISEASE PREDICTION USING DIFFERENT MACHINE LEARNING MODELS

#### Data set feature explanation
###### Sex              |  gender of the patient 
###### Age              |  age of the patient
###### cp               |chest pain 1-Typical Angina 2-Atypical Angina 3-Non-typical Angina
###### trestbps         | Resting blood pressure (in mm Hg on admission to the hospital) 
###### chol             |    serum  cholesterol in mg/dl 
###### fbs              |  (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false) 
###### restecg          |  resting ECG results (0: 'normal', 1 and 2: 'abnormal')
###### thalach          |  maximum heart rate achieved
###### thalach          |  maximum heart rate achieved 
###### exang            |  exercise induced angina (1 = yes; 0 = no)
###### oldpeak          |  ST depression induced by exercise relative to rest
###### slope            |  the slope of the peak exercise ST segment (1: 'upsloping', 2: 'flat', 3: 'down sloping)
###### ca               |  number of major vessels (0-3) colored by fluoroscopy 
###### thal             |  thalassemia (3: 'normal', 6: 'fixed defect' , 7: 'reversible defect')
###### num              |  angiographic disease status (0: no heart disease, > 0: heart disease)

#### Importing Dependencies

```python
import numpy as np # for Array Manipulation and Preprocessing
import pandas as pd # for Data wrangling and preprocessing
import matplotlib.pyplot as plt #for data visualization
import seaborn as sns #for data visualization
```

#### Import Machine Learning Classification Models

```python
## No-Free lunch Theorem 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

#### Importing Evaluation Matrics and Cross Validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score,accuracy_score,precision_score,recall_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading Heart Disease Dataset with contains all the information for the Heart Disease Patient
Data_set = pd.read_csv("heart-disease.csv")
```

```python
#Checking First 10 Rows of the dataset
Data_set.head(10)
```

```python
#Checking for null values
Data_set.isna().sum()
```

```python
# checking statistical relationship between numerical features
Data_set.describe()
```

```python
#Checking Correlation between the features
Data_set.corr()
```

```python
#Visualizing Correlation in heatmap Graph
fig,ax = plt.subplots(figsize=(30,15))
ax = sns.heatmap(Data_set.corr(),annot=True,linewidths=.5,fmt=".2f",cmap="YlGnBu")
```

```python
#Dividing the data into target and features 
X = Data_set.drop(columns="target",axis=1)
Y = Data_set["target"]
```

```python
#Spliting the data into train and test 
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,stratify=Y,random_state=1234)
```

```python
#Checking the shape of the train and test data
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
#Loading the models into a single list name models
models = [KNeighborsClassifier(),SVC(),GaussianNB(),LogisticRegression(),RandomForestClassifier(),DecisionTreeClassifier()]
```

```python
def tr_ts_ac_comp():
    for model in models:
        model.fit(x_train,y_train)
        predicted = model.predict(x_test)
        acc_sco = accuracy_score(y_test,predicted)
        print("The Acuuracy Score for the ",model, " = " ,acc_sco,"For Train Test Split")
```

```python
#Calling the Fuction to see the scores for the different Algorithm   
tr_ts_ac_comp()
```

##### As we see the Logistics Regression has high score and the svc has very low but let split the data into k fold to see the result

```python
def k_fold_split():
    for model in models:
        cv_score = cross_val_score(model,X,Y,cv=5)
        mean_cv_score = sum(cv_score)/ len(cv_score)
        mean_cv_score = mean_cv_score * 100
        mean_cv_score = round(mean_cv_score,2)
        print("The Cross Validation Score for the ",model, "=", cv_score, "and the mean score is ", mean_cv_score)
```

```python
#Calling the kfold function
k_fold_split()
```

##### Again even with the kfold split the Logistic Regression has best mean Score So let try Hyperparameter Tunning to find the best Parameter for the Logistic Regression

```python
## Loading the Logictics Regression Parameter into a Dictionary
params = {
    "penalty":["none","l1","l2"],
    "solver":["lbfgs","liblinear","newton-cg","sag","saga"],
    "C":[1,2,3,4,5]
        

}
```

```python
## Start with Grid Search
GRD = GridSearchCV(LogisticRegression(),params,cv=5)
```

```python
#fitting the data
GRD.fit(X,Y)
```

```python
#Checking for best score
GRD.best_score_
```

```python
#Checking for best parameters
GRD.best_params_
```

```python
#Complete cross validation score and parameters in Dataframe
CV_Results= pd.DataFrame(GRD.cv_results_)
CV_Results.head()
```

```python
GridSerch = CV_Results[["param_C","param_penalty","param_solver"]]
```

```python
GridSerch
```

##### So now we now our best parameter let predict our case study 

```python
model = GRD.best_estimator_
```

```python
prediction = GRD.predict(x_test)
```

```python
accuracy_score(y_test,prediction)
```

##### Making Predictive System
