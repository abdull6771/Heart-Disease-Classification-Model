```python
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
X = Data_set.drop(columns="target",axis=1)
```

```python
Y = Data_set["target"]
```

```python
X
```

```python
Y
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=1234)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
models = [SVC(),KNeighborsClassifier(),GaussianNB(),RandomForestClassifier(),DecisionTreeClassifier()]
```

```python
def comparing():
    for model in models:
        model.fit(x_train,y_train)
        pred = model.predict(x_test)
        accu = accuracy_score(y_test,pred)
        print("The accuracy score of ",model,"is ",accu)
```

```python
comparing()
```

```python
def cross():
    for model in models:
        cv_score = cross_val_score(model,X,Y,cv=5)
        mean_acc = sum(cv_score)/ len(cv_score)
        mean_acc = mean_acc*100
        mean_acc = round(mean_acc,2)
        print("Cross Validation Score for the",model, "=",cv_score)
        print("Mean Accuracy Score for the",model, "=",mean_acc)
```

```python
cross()
```
